```python
#testdaten_wahl = pd.read_csv("./Rohdaten/Testdaten_aus_LTW_2014.csv", encoding="utf-8", sep=";")
testdaten_wahl = pd.read_csv("./Daten_clean/ergebnisse_lt2019.csv", encoding="utf-8", sep=";")
```

```python
testdaten_wahl.info()
```

### Merge mit den historischen Wahldaten

```python
hist_df = pd.read_csv("./Daten_clean/historische_Wahldaten_bereinigt.csv", sep=";", encoding="utf-8")
```

```python
tt = hist_df[hist_df["AGS"]==14524170]
tt
```

```python
test_wahl_aus_lw2014_und_historischen_daten = testdaten_wahl.merge(hist_df, on="AGS", how="left")
test_wahl_aus_lw2014_und_historischen_daten.drop(columns=["Unnamed: 0_x"], inplace=True)
```

```python
test_wahl_aus_lw2014_und_historischen_daten.columns
```

### Richtige Datentypen vergeben - alle Zahlen in np.int64

```python
# alle Ergebnisse in int konvertieren
cols=[i for i in test_wahl_aus_lw2014_und_historischen_daten.columns if i!="WK-Name_LW2014"]
for col in cols:
   
    test_wahl_aus_lw2014_und_historischen_daten[col]=pd.to_numeric(test_wahl_aus_lw2014_und_historischen_daten[col])
    test_wahl_aus_lw2014_und_historischen_daten[col]= test_wahl_aus_lw2014_und_historischen_daten[col].fillna(0).astype(int)
```

```python
test_wahl_aus_lw2014_und_historischen_daten
```

```python
# Ausspielen
test_wahl_aus_lw2014_und_historischen_daten.to_csv("./Daten_clean/ergebnisse_lt_gesamt.csv", 
                                                  encoding="utf-8", sep=";")
```
